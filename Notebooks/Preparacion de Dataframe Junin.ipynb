```python
import pandas as pd
import numpy as np 
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import sklearn 
import seaborn as sb
import datetime
```

```python
df = pd.read_csv("Estimaciones (Provincia de BsAs 1987-2022).csv", encoding='latin-1', sep=";")
df
```

```python
df["Cultivo"].value_counts()
```

```python
#Se eligen las zonas mas cercanas a Junin
df_J= df[df.Departamento.isin(["SALTO", "CHACABUCO","ARENALES","ROJAS","LEANDRO N. ALEM","LINCOLN","JUNIN"])]
df_J
```

```python
df_J["Cultivo"].value_counts()
```

```python
# Se borran las columnas con datos innecesarios
df_J= df_J.drop(['Provincia', 'idProvincia',"Departamento"], axis=1)
df_J
```

```python
#Se eligen los cultivos relevantes 
df_J= df_J[df_J.Cultivo.isin(["Alpiste","Trigo candeal","Cebada cervecera","Arveja", "Avena","Cebada total","Centeno",'Trigo total','Girasol','Maíz','Soja 1ra','Soja 2da','Sorgo'])]
df_J
```

```python
# Se sacan todos los valores invalidos de todas las columnas si los hubiese
df_J['Rendimiento'].replace('SD', '0', inplace=True)
df_J['Sup. Sembrada'].replace('SD', '0', inplace=True)
df_J['Sup. Cosechada'].replace('SD', '0', inplace=True)
df_J['Producción'].replace('SD', '0', inplace=True)
df_J
```

```python
df_J.rename(columns={"Sup. Sembrada":"Sup_Sembrada","Sup. Cosechada":"Sup_Cosechada"}, inplace=True)
```

```python
# Nos aseguramos que todas las 
df_J.Sup_Cosechada = df_J.Sup_Cosechada.astype(float)
df_J.Sup_Sembrada = df_J.Sup_Sembrada.astype(float)
df_J.Producción = df_J.Producción.astype(float)
df_J.Rendimiento = df_J.Rendimiento.astype(float)

df_J
```

```python
anio= lambda x: x[:4]
df_J["anio"]= df_J["Campaña"].astype(str).map(anio)
df_J
```

```python
df_J.anio = df_J.anio.astype(float)
df_J
```

```python
df_J["Cultivo"].unique()
```

```python
#Se pasa a valor codificado la columna Cultivo
df_J=df_J.replace({"Alpiste": 101, "Arveja": 102,'Avena': 103, 'Cebada cervecera': 105, 'Cebada total': 107,
       'Centeno': 109, 'Girasol': 112, 'Maíz': 116, 'Soja 1ra': 124 , 'Soja 2da': 125, 'Sorgo': 127,
       'Trigo candeal': 128, 'Trigo total': 129})
df_J
```

```python
#Se borra la ultima columna objet asi nos quedan todas como float o int
df_J= df_J.drop(["Campaña"], axis=1)
df_J
```
